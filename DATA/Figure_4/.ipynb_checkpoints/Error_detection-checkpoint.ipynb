```python
import numpy as np
import matplotlib.pyplot as plt
```

# Load data

```python
Zero_par_data  = np.load('Zero_par_data.npy')
One_par_data   = np.load('One_par_data.npy')
Plus_par_data  = np.load('Plus_par_data.npy')
Minus_par_data = np.load('Minus_par_data.npy')
Z_values_0_par, P_0_par = Zero_par_data[0], Zero_par_data[1]
Z_values_1_par, P_1_par = One_par_data[0], One_par_data[1]
Z_values_p_par, P_p_par = Plus_par_data[0], Plus_par_data[1]
Z_values_m_par, P_m_par = Minus_par_data[0], Minus_par_data[1]

Zero_pip_data  = np.load('Zero_pip_data.npy')
One_pip_data   = np.load('One_pip_data.npy')
Plus_pip_data  = np.load('Plus_pip_data.npy')
Minus_pip_data = np.load('Minus_pip_data.npy')
Z_values_0_pip, P_0_pip = Zero_pip_data[0], Zero_pip_data[1]
Z_values_1_pip, P_1_pip = One_pip_data[0], One_pip_data[1]
Z_values_p_pip, P_p_pip = Plus_pip_data[0], Plus_pip_data[1]
Z_values_m_pip, P_m_pip = Minus_pip_data[0], Minus_pip_data[1]
```

# Plot data

```python
fig = plt.figure(figsize=(16*9/10,5*9/10), dpi=300)
axs = [fig.add_subplot(241),
       fig.add_subplot(245),
       fig.add_subplot(242),
       fig.add_subplot(246),
       fig.add_subplot(243),
       fig.add_subplot(247),
       fig.add_subplot(244),
       fig.add_subplot(248)]

X = np.arange(1,len(Z_values_0_par)+1)
N = 256*300
N = [256*300 for i in range(10)]+[256*700 for i in range(2)]+[256*1000 for i in range(3)]
from scipy.optimize import curve_fit
def func(x, R, A):
    return -A*np.exp(-x*R)
##################
# ZL plot
##################
# Logical 0
axs[0].errorbar(X*.84+.54, Z_values_0_pip, yerr=np.sqrt(N/np.array(P_0_pip))/N,
                marker='.', linestyle='', label=r'$|0_\mathrm{L}\rangle$ pipelined')
axs[0].errorbar(X*1.1+.54, Z_values_0_par, yerr=np.sqrt(N/np.array(P_0_par))/N,
                marker='.', linestyle='', label=r'$|0_\mathrm{L}\rangle$ parallel')
x=np.linspace(0, 20, 100)
axs[0].plot(x, np.exp(-x/102.2), '--', alpha=.5, color='black', linewidth=1, label='max/min $T_1$')
axs[0].plot(x, np.exp(-x/27.4), '--', alpha=.5, color='black', linewidth=1)
axs[0].fill_between(x, np.exp(-x/102.2), np.exp(-x/27.4), fc='k', alpha=.1)
# Logical 1
axs[2].errorbar(X*.84+.54, Z_values_1_pip, yerr=np.sqrt(N/np.array(P_1_pip))/N,
                marker='.', linestyle='', label=r'$|1_\mathrm{L}\rangle$ pipelined')
axs[2].errorbar(X*1.1+.54, Z_values_1_par, yerr=np.sqrt(N/np.array(P_1_par))/N,
                marker='.', linestyle='', label=r'$|1_\mathrm{L}\rangle$ parallel')
axs[2].plot(x, -np.exp(-x/102.2), '--', alpha=.5, color='black', linewidth=1, label='max/min $T_1$')
axs[2].plot(x, -np.exp(-x/27.4), '--', alpha=.5, color='black', linewidth=1)
axs[2].fill_between(x, -np.exp(-x/102.2), -np.exp(-x/27.4), fc='k', alpha=.1)
# Logical +
axs[4].errorbar(X*.84+.54, Z_values_p_pip, yerr=np.sqrt(N/np.array(P_p_pip))/N,
                marker='.', linestyle='', label=r'$|+_{\! \mathrm{L}}\rangle$ pipelined')
axs[4].errorbar(X*1.1+.54, Z_values_p_par, yerr=np.sqrt(N/np.array(P_p_par))/N,
                marker='.', linestyle='', label=r'$|+_{\! \mathrm{L}}\rangle$ parallel')
axs[4].plot(x, np.exp(-x/116.8), '--', alpha=.5, color='black', linewidth=1, label='max/min $T_2$')
axs[4].plot(x, np.exp(-x/55.1), '--', alpha=.5, color='black', linewidth=1)
axs[4].fill_between(x, np.exp(-x/116.8), np.exp(-x/55.1), fc='k', alpha=.1)
# Logical -
axs[6].errorbar(X*.84+.54, Z_values_m_pip, yerr=np.sqrt(N/np.array(P_m_pip))/N,
                marker='.', linestyle='', label=r'$|-_{\! \mathrm{L}}\rangle$ pipelined')
axs[6].errorbar(X*1.1+.54, Z_values_m_par, yerr=np.sqrt(N/np.array(P_m_par))/N,
                marker='.', linestyle='', label=r'$|-_{\! \mathrm{L}}\rangle$ parallel')
axs[6].plot(x, -np.exp(-x/116.8), '--', alpha=.5, color='black', linewidth=1, label='max/min $T_2$')
axs[6].plot(x, -np.exp(-x/55.1), '--', alpha=.5, color='black', linewidth=1)
axs[6].fill_between(x, -np.exp(-x/116.8), -np.exp(-x/55.1), fc='k', alpha=.1)


for i in [0,4]:
    axs[i].set_ylim(.0, 1.1)
    axs[i].set_yticks([0, .25, 0.5, .75, 1])
    axs[i].set_yticklabels(['', '0.25', '0.5', '0.75', '1'])
    axs[i].legend(loc='lower left', frameon=False)
for i in [2,6]:
    axs[i].set_ylim(-1.1, .0)
    axs[i].set_yticks([0, -.25, -0.5, -.75, -1])
    axs[i].set_yticklabels(['', '-0.25', '-0.5', '-0.75', '-1'])
    axs[i].legend(loc='upper left', frameon=False)
for i in [0,2]:
    axs[i].set_ylabel(r'Logical Z, $\langle Z_\mathrm{L} \rangle$')
    axs[i].set_xlim(0, 18)
    axs[i].set_xlabel('Time ($\mathrm{\mu s}$)')
    axs[i].tick_params(axis='both', direction='in', which='both')
for i in [4,6]:
    axs[i].set_ylabel(r'Logical X, $\langle X_\mathrm{L} \rangle$')
    axs[i].set_xlim(0, 18)
    axs[i].set_xlabel('Time ($\mathrm{\mu s}$)')
    axs[i].tick_params(axis='both', direction='in', which='both')

##################
# Error plot
##################
def func(x, R, A):
    return A*np.exp(-x*R)

def func2(x, R, A):
    return np.log(A)-R*x

for P_par, P_pip, i in zip([P_0_par, P_1_par, P_p_par, P_m_par],
                           [P_0_pip, P_1_pip, P_p_pip, P_m_pip],
                           [1, 3, 5, 7]):
    popt, pcov = curve_fit(func2, X, np.log(P_pip))
    perr = np.sqrt(np.diag(pcov))
    R_pip = popt[0]
    axs[i].plot(X, func(X, *popt), 'C0--', linewidth=1)
    axs[i].plot(X, P_pip, 'C0.', label='pipelined')
    popt, pcov = curve_fit(func2, X, np.log(P_par))
    R_par = popt[0]
    perr = np.sqrt(np.diag(pcov))
    axs[i].plot(X, func(X, *popt), 'C1--', linewidth=1)
    axs[i].plot(X, P_par, 'C1.', label='parallel')

    axs[i].set_ylim(1e-4, 1)
    axs[i].set_xlabel('Error detection cycles, $n$')
    axs[i].set_ylabel(r'Post-selected fraction, $P$')
    axs[i].set_yscale('log')
    axs[i].set_xticks([0,5,10,15])
    axs[i].legend(loc='lower left', bbox_to_anchor=(.02, 0.02), frameon=False)
    axs[i].tick_params(axis='both', direction='in', which='both')

fig.tight_layout()
```
