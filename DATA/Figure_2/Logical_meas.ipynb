```python
import numpy as np
import matplotlib.pyplot as plt
```

# Load data

```python
# States on the equator
Phi_data = np.load('Phi_data.npy')
ZL_phi = Phi_data[0]
XL_phi = Phi_data[1]
YL_phi = Phi_data[2]
PL_Z_phi = Phi_data[3]
PL_X_phi = Phi_data[4]
PL_Y_phi = Phi_data[5]
# States on the X-Z plane
Theta_data = np.load('Theta_data.npy')
ZL_theta = Theta_data[0]
XL_theta = Theta_data[1]
YL_theta = Theta_data[2]
PL_Z_theta = Theta_data[3]
PL_X_theta = Theta_data[4]
PL_Y_theta = Theta_data[5]
# Fidelities
FL_theta = [0.999091466021245, 0.9981820682063176, 0.9943984188935455, 0.9973250160354172, 0.9907425849870977, 0.9829795628201901, 0.9802408959292251, 0.9722767281510792, 0.9648626241085394, 0.9748040002562975, 0.9781840397816516, 0.9816084566264731, 0.9819427369235308, 0.9925752774881004, 0.992598494336439, 0.9921935613694874, 0.9972440947482841]
FL_phi   = [0.9775629522041632, 0.9701112795887342, 0.9728231287537918, 0.9635177981394291, 0.9675953201471861, 0.970413416845392, 0.952090463328645, 0.9538268208919619, 0.9439960454838253, 0.9632208427803555, 0.9514873202943992, 0.9620320543898102, 0.9615351580821766, 0.9607505238615877, 0.9534418251735226, 0.9570099352471308, 0.9554006509773635, 0.957732388315429, 0.9685473291165709, 0.9731012332220281, 0.9509349098660297, 0.9603455252695803, 0.9466045096752576, 0.9610395755572677, 0.9519788662865128, 0.9544159627007683, 0.954680894644576, 0.9627361097453445, 0.9561091066456373, 0.95643243397686, 0.979753583168309, 0.9592771808702214, 0.9682431448154571]
```

# States in the equator

## Fit model

```python
from scipy.optimize import curve_fit

def func_x(theta, amp):
    return amp*np.cos(np.pi/180*theta)

def func_y(theta, amp):
    return amp*np.sin(np.pi/180*theta)

x_data = np.linspace(0, 360, 33)
popt_y, pcov_y = curve_fit(func_y, x_data, YL_phi)
popt_x, pcov_x = curve_fit(func_x, x_data, XL_phi)
```

## Plot data

```python
fig= plt.figure(figsize=(3.5, 5), dpi=200)
axs = [fig.add_subplot(211),
       fig.add_subplot(212)]

#############################
# Logical expectation values
#############################
axs[0].axhline(y=1, color='black', linestyle='--', alpha=.25)
axs[0].axhline(y=-1, color='black', linestyle='--', alpha=.25)

Phi_p = np.linspace(0, 360, 361)
Phi = np.linspace(0, 360, 33)
axs[0].plot(Phi_p, func_y(Phi_p, *popt_y), 'C3--')
axs[0].plot(Phi_p, func_x(Phi_p, *popt_x), 'C0--')
axs[0].plot(Phi_p, np.zeros(len(Phi_p)), 'C2--')
axs[0].plot(Phi, ZL_phi, 'C2o', label=r'$\langle Z_\mathrm{L}\rangle$', markersize=4)
axs[0].plot(Phi, XL_phi, 'C0o', label=r'$\langle X_\mathrm{L}\rangle$', markersize=4)
axs[0].plot(Phi, YL_phi, 'C3o', label=r'$\langle Y_\mathrm{L}\rangle$', markersize=4)
axs[0].plot(Phi, 2*np.array(FL_phi)-1, 'k-', lw=1, label=r'$2F_\mathrm{L}-1$')
axs[0].plot(Phi,-2*np.array(FL_phi)+1, 'k-', lw=1)

axs[0].set_xticks([0, 90, 180, 270, 360])
axs[0].set_xlabel(r'Rotation angle, $\phi$')
axs[0].set_yticks([-1, -.5, 0, .5, 1])
axs[0].set_yticklabels(['-1', '', '0', '', '1'])
axs[0].set_ylabel(r'Logical $\hat{O}$ , $\langle \hat{O}_\mathrm{L}\rangle$')
axs[0].tick_params(axis='both', direction='in', which='both')
axs[0].legend(frameon=False, fontsize=8)

#############################
# Logical expectation values
#############################
axs[1].plot(Phi, PL_Z_phi, 'C2-', label=r'$\langle Z_\mathrm{L}\rangle$', markersize=4)
axs[1].plot(Phi, PL_X_phi, 'C0-', label=r'$\langle X_\mathrm{L}\rangle$', markersize=4)
axs[1].plot(Phi, PL_Y_phi, 'C3-', label=r'$\langle Y_\mathrm{L}\rangle$', markersize=4)

axs[1].set_xticks([0, 90, 180, 270, 360])
axs[1].set_xlabel(r'Rotation angle, $\phi$')
axs[1].set_ylim(0,.4)
axs[1].set_ylabel(r'P-S fraction, $P$')
axs[1].tick_params(axis='both', direction='in', which='both')
axs[1].legend(frameon=False)

fig.tight_layout()
```

# States in the X-Z plane

## Fit model

```python
from scipy.optimize import curve_fit

def func_z(theta, amp):
    return amp*np.cos(np.pi/180*theta)*(1 + .5*np.sin(np.pi/180*theta)**2)

def func_x(theta, amp):
    return amp*.5*np.sin(np.pi/180*theta)**2/(1 - .5*np.sin(np.pi/180*theta)**2)

x_data = np.linspace(0, 180, 17)
popt_z, pcov_z = curve_fit(func_z, x_data, ZL_theta[:17])
popt_x, pcov_x = curve_fit(func_x, x_data, XL_theta[:17])
```

## Plot data

```python
fig= plt.figure(figsize=(3.5, 5), dpi=200)
axs = [fig.add_subplot(211),
       fig.add_subplot(212)]

#############################
# Logical expectation values
#############################
axs[0].axhline(y=1, color='black', linestyle='--', alpha=.25)
axs[0].axhline(y=-1, color='black', linestyle='--', alpha=.25)

Theta_p = np.linspace(0, 180, 181)
Theta = np.linspace(0, 180, 17)
axs[0].plot(Theta_p, func_z(Theta_p, *popt_z), 'C2--')
axs[0].plot(Theta_p, func_x(Theta_p, *popt_x), 'C0--')
axs[0].plot(Theta_p, np.zeros(len(Theta_p)), 'C3--')
axs[0].plot(Theta, ZL_theta[:17], 'C2o', label=r'$\langle Z_\mathrm{L}\rangle$', markersize=4)
axs[0].plot(Theta, XL_theta[:17], 'C0o', label=r'$\langle X_\mathrm{L}\rangle$', markersize=4)
axs[0].plot(Theta, YL_theta[:17], 'C3o', label=r'$\langle Y_\mathrm{L}\rangle$', markersize=4)
axs[0].plot(Theta, 2*np.array(FL_theta)-1, 'k-', lw=1, label=r'$2F_\mathrm{L}-1$')
axs[0].plot(Theta,-2*np.array(FL_theta)+1, 'k-', lw=1)

axs[0].set_xticks([0, 45, 90, 135, 180])
axs[0].set_xlabel(r'Rotation angle, $\theta$')
axs[0].set_yticks([-1, -.5, 0, .5, 1])
axs[0].set_yticklabels(['-1', '', '0', '', '1'])
axs[0].set_ylabel(r'Logical $\hat{O}$ , $\langle \hat{O}_\mathrm{L}\rangle$')
axs[0].tick_params(axis='both', direction='in', which='both')
axs[0].legend(frameon=False, fontsize=8)

#############################
# Logical expectation values
#############################
axs[1].plot(Theta, PL_Z_theta[:17], 'C2-', label=r'$\langle Z_\mathrm{L}\rangle$', markersize=4)
axs[1].plot(Theta, PL_X_theta[:17], 'C0-', label=r'$\langle X_\mathrm{L}\rangle$', markersize=4)
axs[1].plot(Theta, PL_Y_theta[:17], 'C3-', label=r'$\langle Y_\mathrm{L}\rangle$', markersize=4)

axs[1].set_xticks([0, 45, 90, 135, 180])
axs[1].set_xlabel(r'Rotation angle, $\theta$')
axs[1].set_ylim(0,.4)
axs[1].set_ylabel(r'P-S fraction, $P$')
axs[1].tick_params(axis='both', direction='in', which='both')
axs[1].legend(frameon=False)

fig.tight_layout()
```
